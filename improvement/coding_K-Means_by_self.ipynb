```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn import datasets, preprocessing
import random, sys
```

```python
# load data
iris_set = datasets.load_iris()
iris_data = iris_set.data
iris_target = iris_set.target

print('iris_data=>'+str(iris_data))
print('iris_target=>'+str(iris_target))

min_max_scaler = preprocessing.MinMaxScaler()
iris_set_normal = min_max_scaler.fit_transform(iris_data)
print('iris_normal=>'+str(iris_set_normal))
```

```python
# KMeans by Self
def my_kmeans(train_data, k, train_target):
    r_n = len(train_data)
    f_n = len(train_data[0])
    n_k_min = r_n / k / 4 # 类别内个数少于此时，重新随机

    core_list = [[random.random() for j in range(f_n)] for i in range(k)]
    print('core_list=>'+str(core_list))
    for eps in range(20): # 迭代次数
        row_core=[] # 当前行的最近类（中心）
        for i in range(r_n): # 遍历数据记录
            core_dist=[] # 当前行，每个
            for j in range(k): # 遍历每个类中心
                t_dist = 0.0
                for m in range(f_n): # 遍历每个特征
                    t_dist=t_dist+ abs(core_list[j][m] - train_data[i][m])
                core_dist.append(t_dist) # 当前类别的距离
            i_k = np.argmin(core_dist) # 最近的类别序号
            row_core.append(i_k) # 当前行的类别

        core_f_sum = [[0 for j in range(f_n)] for i in range(k)] # 各分类各指标
        core_row_cnt = [0 for i in range(k)]  # 各分类的个数
        for i in range(r_n):  # 遍历数据记录
            core_idx = row_core[i]
            core_row_cnt[core_idx] =+1
            for m in range(f_n): # 遍历每个特征,做同类的累加
                core_f_sum[core_idx][m] =+ train_data[i][m]


        # 更新类别中心位置
        for j in range(k): # 遍历每个类中心
            n_k = core_row_cnt[j]
            if n_k>n_k_min:
                for m in range(f_n): # 遍历每个特征,求均值
                    core_list[j][m] = core_f_sum[j][m] / n_k
            else:
                # 类内个数小于阈值则重新随机
                core_list[j] = [random.random() for f in range(f_n)]

        for i in range(r_n):  # 遍历数据记录
            print(str(eps*10000+i)+'=>'+str(train_data[i])+' -> '+str(row_core[i])+' <- '+str(train_target[i]))

    return train_data, row_core, core_list #返回各类的中心位置
```

```python
# Use KMeans by Self
_, y_pred_1, _ = my_kmeans(iris_set_normal, 3, iris_target)
```

```python
# KMeans by Sklearn
def sklearn_kmeans(train_data, k, train_target):
    r_n = len(train_data)
    model = KMeans(k, init='random')

    result = model.fit(train_data)
    labels = result.labels_


    for i in range(r_n):  # 遍历数据记录
        print(str(210000 + i) + '=>' + str(train_data[i]) + ' -> ' + str(labels[i]) + ' <- ' + str(train_target[i]))

    return train_data, labels
```

```python
# Use K-Means by Sklearn
_, y_pred_2= sklearn_kmeans(iris_set_normal, 3, iris_target)
```

```python
# The accuracy of K-Means by Self
(y_pred_1 == iris_target).sum() / len(iris_target)
```

```python
# The accuracy of K-Means by Sklearn
(y_pred_2 == iris_target).sum() / len(iris_target)
```
