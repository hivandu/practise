```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0, 5, 0.02)
y = np.exp(-x) * np.cos(2*np.pi*x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid(color = 'b')
```

# ax.grid(b=None, which='major', axis='both', **kwargs)
    """[summary]
    b:
        None: 不显示，并且kwargs不接收任何参数
        True: 显示
        False: 不限时，但是kwargs接收参数
    which:
        'major'：表示显示主刻线的坐标网格
        'minor'：表示显示副刻线的坐标网格。当然，对于上面的示例而言，只有主刻线
        'both'：以上都显示
    axis:
        both: 显示横纵轴
        x: 显示纵轴
        y: 显示横轴
    **kwargs:
        color: 颜色
        linestyle: `-`,`--`,`-.`, `:`
        linewidth: int
    """

```python
x = np.arange(0, 5, 0.02)
y = np.exp(-x) * np.cos(2*np.pi*x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid(b=False, which='both', axis='x', color='r', linestyle=':', linewidth='2')
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlim((-1, 6)) # x轴范围
ax.set_ylim((-1, 2)) # y轴范围
```

```python
# 标记坐标轴
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlim(left = 10, right = -2)
ax.set_xlabel('axis x')
ax.set_ylabel('axis y')
```

```python
# 调整字号
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_xlim(left = 10, right = -2)
ax.set_xlabel('axis x', fontsize=20, color='r') # 设置字号和颜色
ax.set_ylabel('axis y', fontsize = 'xx-large') # 另外一种方式
ax.set_title('The func', fontdict={'fontsize':22, 'fontweight':'medium'})
```

```python
# 设置刻度
fig, ax = plt.subplots()
ax.plot(x, y)
ax.yaxis.set_major_locator(plt.NullLocator()) 
ax.xaxis.set_major_formatter(plt.NullFormatter())
```

- NullLocator：秉承着 “望文生义” 的原则，这是很重要的命名原则，估计已经猜测到这个函数的作用了。对的，它的作用就是让刻线消失，即返回无刻线对象；既然无刻线了，那么标示也就没有了，正所谓“皮之不存毛将焉附”。因此，这个函数的作用就在于，如果想要把某个坐标轴的刻度都取消了，可以用它的返回值为对象。
- NullFormatter：对照着上面的，理解这个函数就比较简单了，它返回的是没有标示的对象。
- set_major_locator：设置主刻度的刻线。
- set_major_formatter：设置主刻度的标示。

```python
# 🌰
from sklearn.datasets import fetch_olivetti_faces

faces = fetch_olivetti_faces().images
```

```python
fig, ax = plt.subplots(5,5,figsize = (5,5))
fig.subplots_adjust(hspace=0,wspace=0)

for i in range(5):
    for j in range(5):
        ax[i,j].xaxis.set_major_locator(plt.NullLocator())
        ax[i,j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10*i+j], cmap='bone')
```

```python
import numpy as np

x = np.linspace(0,10,50)
y1 = np.exp(x)
y2 = np.sin(x)

# (x, y1)
fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'b')
ax1.set_xlabel('x axis')
ax1.set_ylabel('exp', color = 'b')
ax1.tick_params(axis = 'y', which='major', colors = 'b', width=4, length = 9)

# (x, y2)
ax2 = ax1.twinx()
ax2.plot(x, y2, color = 'r')
ax2.set_ylabel('sin', color = 'r')
ax2.tick_params(axis = 'y', colors = 'r', direction='inout')
```

```python
# 调节
import sys
import numpy as np
from matplotlib.ticker import AutoMinorLocator
x = np.linspace(0, 10, 50)
y1 = np.exp(x)
y2 = np.sin(x)

fig, ax1 = plt.subplots()

ax1.plot(x, y1, 'b')
ax1.set_label('x axis')
ax1.set_xlim(0, 10)
ax1.set_xticks(range(11))
ax1.set_ylabel('exp', color = 'b')
ax1.set_ylim(0, 3000)
ax1.set_yticks(range(0, 3001, 600))
ax1.yaxis.set_minor_locator(AutoMinorLocator(5))
ax1.tick_params(axis = 'y', which = 'major', colors = 'b', direction = 'in')

ax2 = ax1.twinx()
ax2.plot(x, y2, color = 'r')
ax2.set_ylabel('sin', color = 'r')
ax2.tick_params(axis='y', colors = 'r', direction = 'inout')
```
