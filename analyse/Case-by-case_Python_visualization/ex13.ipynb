```python
## 饼图
```

```python
import pandas as pd
import numpy as np
import plotly as py
import plotly.graph_objs as go
py.offline.init_notebook_mode(connected=True)
```

```python
dataSet = pd.read_csv('~/data/cbcpv/universityrank/timesData.csv')
dataSet.sample(5)
```

```python
dfcn = dataSet[dataSet['country'] == 'China']
dfcn5 = dfcn.iloc[:5, :]
dfcn5
```

```python
dfcn5.info()
```

```python
dfcn5['num_students'] = dfcn5['num_students'].apply(lambda value: float(value.replace(',', '')))
dfcn5['num_students']
```

```python
trace=go.Pie(
    values=dfcn5['num_students'],
    labels=dfcn5['university_name'],
    domain={'x':[0, .5]},
    hoverinfo='label + percent + name',
    hole = 0
)
layout=go.Layout(
    title='Universities Number of Students Rates',
    annotations=[
        {
            'font': {'size': 20},
            'showarrow': False,
            'text': 'Number of Students',
            'x': 0.20,
            'y': 1
        }
    ]
)
fig = go.Figure(data=[trace], layout=layout)
py.offline.iplot(fig)
```

```python
colors = ['#FEBFB3', '#E1396C', '#96D38C', '#D0F9B1', '##CD5C5C']
trace=go.Pie(
    values=dfcn5['num_students'],
    labels=dfcn5['university_name'],
    rotation=45,
    opacity=.8,
    showlegend=False,
    pull=[0.1,0,0,0,0],
    hoverinfo='label+percent',
    textinfo='value',
    textfont=dict(size=20, color='white'),
    marker=dict(colors=colors, line=dict(color='#333333', width=2))
)

py.offline.iplot([trace])
```

```python
# 获取数据集中美国的TOP 5 学校
dfusa=dataSet[dataSet['country'] == 'United States of America']
dfusa5=dfusa.iloc[:5, :]
dfusa5['num_students']=dfusa5['num_students'].apply(lambda value: float(value.replace(',','')))

trace0=go.Pie(
    values=dfcn5['num_students'],
    labels=dfcn5['university_name'],
    domain=dict(x=[0, .45]),
    name='China Top5',
    showlegend=False,
    hoverinfo='label + percent + name',
    textinfo='percent',
    hole=.4
)

trace1=go.Pie(
    values=dfusa5['num_students'],
    labels=dfusa5['university_name'],
    domain=dict(x=[0.55, 1]),
    name='USA Top5',
    showlegend=False,
    hoverinfo='label + percent + name',
    textinfo='percent',
    hole=.4
)
data = go.Data([trace0, trace1])

layout=go.Layout(
    title='TOP5 of China and USA',
    annotations=[
        dict(
            font={'size':20},
            text='China',
            x=.23,
            y=.4
        ),
        dict(
            font={'size':20},
            showarrow=False,
            text='USA',
            x=.81,
            y=.5
        )
    ]
)

fig=go.Figure(data=data, layout=layout)
py.offline.iplot(fig)
```

```python
import plotly as py
import plotly.graph_objs as go
import numpy as np

x = np.random.randn(500)
data = [go.Histogram(x=x, histnorm='probability')]
py.offline.iplot(data, filename='normalized histogram')
```

```python
x2011 = dataSet.student_staff_ratio[dataSet.year==2011]
x2012 = dataSet.student_staff_ratio[dataSet.year==2012]
```

```python
trace1=go.Histogram(
    x=x2011,
    opacity=.8,
    name='2011',
    marker=dict(color='rgba(171,50,86,.6)')
)
trace2=go.Histogram(
    x=x2012,
    opacity=.75,
    name='2012',
    marker=dict(color='rgba(12,50,196,.6)')
)

data=[trace1, trace2]

layout=go.Layout(
    barmode='overlay',
    title='students-stff ratio in 2011 and 2012',
    xaxis=dict(title='students-staff ratio'),
    yaxis=dict(title='Cont')
)
fig=go.Figure(data=data, layout=layout)
py.offline.iplot(fig)
```

```python
## 统计图的布局
```

```python
### 嵌入式
```

```python
df100=dataSet.iloc[:100, :]
# 主图trace对象
trace0 = go.Scatter(
    x=df100['world_rank'],
    y=df100['teaching'],
    name='teaching',
    marker=dict(color='rgba(16,112,2,0.8)')
)
trace1=go.Scatter(
    x=dataSet['world_rank'],
    y=dataSet['income'],
    xaxis='x2',
    yaxis='y2',
    name='income',
    marker=dict(color='rgba(6,6,6,.8)')
)

data = go.Data([trace0, trace1])
layout=go.Layout(
    xaxis2=dict(
        domain=[0.6, 0.95],
        anchor='y2'
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],
        anchor='x2'
    ),
    title='Income and Teaching vs World Rank of Universities'
)

fig=go.Figure(data=data, layout=layout)
py.offline.iplot(fig)
```

```python
### 分区
```

```python
trace1=go.Scatter(
    x=df100.world_rank,
    y=df100.research,
    name='research'
)
trace2=go.Scatter(
    x=df100.world_rank,
    y=df100.citations,
    xaxis='x2',
    yaxis='y2',
    name='citations'
)
trace3=go.Scatter(
    x=dataSet.world_rank,
    y=dataSet.income,
    xaxis='x3',
    yaxis='y3',
    name='income'
)
trace4=go.Scatter(
    x=dataSet.world_rank,
    y=dataSet.total_score,
    xaxis='x4',
    yaxis='y4',
    name='total_score'
)

data = [trace1, trace2, trace3, trace4]
layout=go.Layout(
    xaxis=dict(domain=[0, .45]), # 3
    yaxis=dict(domain=[0, .45]), # 4
    xaxis2=dict(domain=[0.55, 1]), # 5
    xaxis3=dict(domain=[0, .45], anchor='y3'), # 6
    xaxis4=dict(domain=[0.55,1], anchor='y4'),
    yaxis2=dict(domain=[0, .45], anchor='x2'), # 7
    yaxis3=dict(domain=[.55,1]),
    yaxis4=dict(domain=[.55,1], anchor='x4'),
    title='Research, citation, income and total score VS World Rank of Universities'
)

fig=go.Figure(data=data, layout=layout)
py.offline.iplot(fig)

```
