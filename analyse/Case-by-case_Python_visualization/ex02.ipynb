## 多样化的图像

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 10, 0.005)
y = np.exp(-x/2) * np.sin(2 * np.pi * x)
fig, ax = plt.subplots()
ax.plot(x, y)

# 设置坐标区间
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)

# 移动坐标轴
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position('zero')

# 以下标注
x1 = 1.25
y1 = np.exp(-x1/2)*np.sin(2*np.pi*x1)

# 呈现图中的坐标点
ax.scatter([x1,],[y1,], 50, color='b')
# 从坐标点 (x1, y1) 分别向 x 轴、y 轴画出垂直的虚线
ax.plot([x1,x1],[0,y1], color='b', linewidth=2.5, linestyle='--')
ax.plot([0, x1],[y1, y1], color='b', linewidth=2.5, linestyle='--')


# 标注文字说明
ax.annotate(r'$e^{-\frac{x}{2}}sin(2\pi{x}), x=1.25$', xy=(x1, y1), xycoords='data', xytext=(+30, +0.6), textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))
```

```python
import pandas as pd
cities = pd.read_csv('~/data/cbcpv/jiangsu/city_population.csv')
cities
```

```python
lat = cities['latd']
lon = cities['longd']

population = cities['population']
area = cities['area']
np.log10(population)
```

```python
plt.rcParams['font.sans-serif']  = ['Arial Unicode MS'] # Chinese
plt.scatter(lon, lat, label=None, c=np.log10(population), cmap="viridis", s=area/10, linewidths=0, alpha=0.5)   

plt.axis('equal')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')

for area in [10, 30, 50]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')

plt.legend(scatterpoints=1, frameon = False, labelspacing=1, title='City Area')
plt.title('江苏各城市面积和人口')
```

```python
np.random.seed(19680801)

x = np.arange(0.0, 50.0, 2.0)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0
s = np.random.rand(*x.shape) * 800 + 500

plt.scatter(x, y, s, c='g', alpha=.5, marker=r'$\clubsuit$', label='Luck')

plt.xlabel('Leprechauns')
plt.ylabel('Gold')
plt.legend(loc='upper left')
```

```python
np.random.seed(19680801)

# 点的面积和色彩
N = 150
r = 2 * np.random.rand(N)
theta = 2*np.pi * np.random.rand(N)
area = 200 * r**2
colors = theta

fig=plt.figure()
ax = fig.add_subplot(111, projection='polar')
ax.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=0.75)
```
