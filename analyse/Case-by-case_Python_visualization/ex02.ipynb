## 多样化的图像

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 10, 0.005)
y = np.exp(-x/2) * np.sin(2 * np.pi * x)
fig, ax = plt.subplots()
ax.plot(x, y)

# 设置坐标区间
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)

# 移动坐标轴
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position('zero')

# 以下标注
x1 = 1.25
y1 = np.exp(-x1/2)*np.sin(2*np.pi*x1)

# 呈现图中的坐标点
ax.scatter([x1,],[y1,], 50, color='b')
# 从坐标点 (x1, y1) 分别向 x 轴、y 轴画出垂直的虚线
ax.plot([x1,x1],[0,y1], color='b', linewidth=2.5, linestyle='--')
ax.plot([0, x1],[y1, y1], color='b', linewidth=2.5, linestyle='--')


# 标注文字说明
ax.annotate(r'$e^{-\frac{x}{2}}sin(2\pi{x}), x=1.25$', xy=(x1, y1), xycoords='data', xytext=(+30, +0.6), textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))
```
