```python
%matplotlib inline
import plotnine as pn
from plotnine import data
base_plot = pn.ggplot(data.mpg, pn.aes(
    x='displ', 
    y='hwy',
    color='factor(cyl)'
))
base_plot + pn.geom_point()
```

```python
%matplotlib inline
import plotnine as pn
from plotnine import data

(pn.ggplot(data.mpg)
    + pn.aes(
        x='displ',
        y='hwy'
    )
    + pn.geom_point(pn.aes(
        color='factor(cyl)'
    ))
    + pn.geom_smooth()
)
```

```python
# # geom_point 完整参数
geom_point(
    mapping=None,
    data=None,
    stat='identity',
    position='identity',
    na_rm=False,
    inherit_aes=True,
    show_legend=None,
    **kwargs
)
```

```python
(pn.ggplot(data.mpg)
    + pn.aes(x='displ', y='hwy')
    + pn.geom_point(pn.aes(
        color='factor(cyl)',
        size='hwy'
    ))
    + pn.geom_smooth(color='red')
)
```

```python
(pn.ggplot(data.mpg)
    + pn.aes(x='displ',y='hwy')
    + pn.geom_point(
        pn.aes(
            color='factor(cyl)',
            size='hwy'
        )
    )
    + pn.geom_smooth(color='red')
    + pn.xlab('engine displacement(L)')
)
```

```python
(pn.ggplot(data.mpg)
    + pn.aes(x='displ',y='hwy')
    + pn.geom_point(
        pn.aes(
            color='factor(cyl)',
            size='hwy'
        )
    )
    + pn.geom_smooth(color='red')
    + pn.xlab('engine displacement(L)')
    + pn.scale_x_reverse()
)
```

```python
# 创建数据集
import pandas as pd
c_chord = pd.DataFrame({
    'Fret': [0, 2.5, 1.5, 0, 0.5, 0],
    'String': [1,2,3,4,5,6]
})
c_chord['Sequence'] = list(range(1, 1+len(c_chord['Fret'])))
c_chord
```

```python
markings = pd.DataFrame({
    'Fret':   [2.5, 4.5, 6.5, 8.5, 11.5, 11.5, 14.5, 16.5, 18.5, 20.5],     
    'String': [3.5, 3.5, 3.5, 3.5, 2, 5, 3.5, 3.5, 3.5, 3.5]
})
markings
```

```python
(pn.ggplot(
    c_chord, 
    pn.aes('Fret', 'String')
))
```

```python
neck_color = '#FFDDCC'
fret_color = '#998888'
string_color = '#AA9944'

neck_theme = pn.theme(
    figure_size=(10, 2),
    panel_background = pn.element_rect(fill=neck_color),
    panel_grid_major_y = pn.element_line(color=string_color, size=2.2),
    panel_grid_major_x = pn.element_line(color=fret_color, size=2.2),
    panel_grid_minor_x = pn.element_line(color=fret_color, size=1)
)
(pn.ggplot(c_chord, pn.aes('Fret', 'String'))
    + neck_theme
)
```

```python
neck_color = '#FFDDCC'
fret_color = '#998888'
string_color = '#AA9944'

neck_theme = pn.theme(
    figure_size=(10, 2), # 定义图像大小
    panel_background = pn.element_rect(fill=neck_color), # 定义图的颜色
    
    # 定义Y轴主刻度网格线
    panel_grid_major_y = pn.element_line(color=string_color, size=2.2),
    panel_grid_major_x = pn.element_line(color=fret_color, size=2.2),
    panel_grid_minor_x = pn.element_line(color=fret_color, size=1)
)
(pn.ggplot(c_chord, pn.aes('Fret', 'String'))
    + neck_theme # 把定义主题当作一个对象
    + pn.geom_point( # 绘制散点图
        pn.aes(color='Sequence'),
        fill='#FFFFFF',
        size=3
    )
    + pn.geom_path( # 将各个散点图用折线链接起来
        pn.aes(color='Sequence'),
        size=3
    )
)
```
