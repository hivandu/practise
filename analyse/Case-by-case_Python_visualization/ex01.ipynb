```python
# 重叠坐标轴
import numpy as np
# 生成两组数据集
x = np.linspace(0, 10, 50)
y1 = np.exp(x)
y2 = np.sin(x)

# 绘制(x, y1)
# 创建坐标系对象 ax1,然后在这个对象中绘制(x, y1)这组数据的图像,即结果中呈现的蓝色图线。
fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'b')
ax1.set_xlabel('axis x', fontsize = 20)
ax1.set_ylabel('exp', color = 'b', fontsize=22)
# 对 ax1 坐标系的 y 轴进行设置
ax1.tick_params(axis='y', which='major', color='b', width=4, length=9)

# 绘制(x, y2)
# 将另外一个坐标系可以视为一个新的图层,叠加到原来坐标系之上
ax2 = ax1.twinx()
ax2.plot(x, y2, 'r')
ax2.set_ylabel('sin', color='r', fontsize=22)
ax2.tick_params(axis='y', color='r', direction='inout')
```

```python
# 精细化处理上面代码
import sys
import numpy as np
from matplotlib.ticker import AutoMinorLocator

x = np.linspace(0, 10, 50)
y1 = np.exp(x)
y2 = np.sin(x)

fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'b')
ax1.set_xlabel('axis x',fontsize=20)
ax1.set_xlim(0, 10)
ax1.set_xticks(range(11))
ax1.set_ylabel('exp',color='b',fontsize=22)
# 设置了纵轴显示的数值范围
ax1.set_ylim(0, 3000)
# 设置了主刻度的标示所显示的数值
ax1.set_yticks(range(0, 3001, 600))
# 设置副刻度的刻线的,并且以 AutoMinorLocator(5) 规定了主刻线之间的副刻线的段数
ax1.yaxis.set_minor_locator(AutoMinorLocator(5))
ax1.tick_params(axis='y', which='major', colors = 'b', direction='in')

ax2 = ax1.twinx()
ax2.plot(x, y2, color='r')
ax2.set_ylabel('sin', color = 'r', fontsize=22)
ax2.tick_params(axis='y', color='r', direction = 'inout')
```

```python
## GridSpec

# 引入 GridSpec,注意它不在 plt 下面
from matplotlib.gridspec import GridSpec
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS'] # Chinese
fig = plt.figure()
fig.set_facecolor('w') # bg
# 使用图像对象的方法设置本图像的标题
fig.suptitle('划分不同大小的坐标系区分')

# 建立了 GridSpec 的实例
gs = GridSpec(3, 3, figure=fig)
# GridSpec(nrows, ncols, figure=None, left=None, bottom=None, right=None, top=None, wspace=None, hspace=None, width_ratios =None, height_ratios=None)

# fig.add_subplot 方法旨在创建一个坐标系分区,其效果与 add_axes 方法相仿
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, :-1])
ax3 = fig.add_subplot(gs[1:, -1])
ax4 = fig.add_subplot(gs[-1, 0])
ax5 = fig.add_subplot(gs[-1, -2])

def format_axes(fig):
    # fig.axes 是通过图像对象的属性 axes 得到该图像中的所有坐标系对象,然后依次调用每个坐标系对象,使用坐标系对象的属性和方法
    for i, ax in enumerate(fig.axes):
        ax.text(0.5, 0.5, "ax%d" %(i+1), va='center', ha='center')
        ax.tick_params(labelbottom=False, labelleft=False)

format_axes(fig)
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color='b', linewidth=2.5, linestyle='-')
plt.plot(X,S, color='g', linewidth=1.0, linestyle='-.')

# plt.xticks 和 ax.set_xticks是不同应用模式下的同一个功能
# 除了规定横坐标的主刻线位置（[-np.pi, -np.pi/2, 0, np.pi/2, np.pi]）之外，还规定了主刻线的标示显示方式（[r'−π', r'−π/2', r'0', r'+π/2', r'+π']）
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.yticks([-1,0,+1], [r'$-1$', r'$0$', r'$+1$'])

plt.tick_params(axis='both', labelsize=18, colors='r')
```

```python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
c,s = np.cos(x), np.sin(x)

plt.plot(x, c, color='b', linewidth=2.5, linestyle='-')
plt.plot(x, s, color='r', linewidth=1.0, linestyle='-.')

plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$',r'$-\pi/2$',r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, +1], [r'$-1$', r'$0$', r'$+1$'])

plt.tick_params(axis='both', labelsize=18, colors="r")

# 增加内容
# plt.gca 函数的作用是获得当前的坐标系对象
ax = plt.gca()
# ax.spines 是坐标系对象四周的边线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
# ax.xaxis 显然是获得 x 轴,前面已经明确过了, 坐标轴上有刻度(包含刻线和标示)
ax.xaxis.set_ticks_position('bottom')
# 移动下边线(ax.spines['bottom']), 把它作为 x 轴,移动到 y=0 的位置
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
```
