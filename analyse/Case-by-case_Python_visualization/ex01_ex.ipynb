```python
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces().images

fig, ax = plt.subplots(5, 5, figsize=(5,5))
fig.subplots_adjust(hspace=0, wspace=0)

for i in range(5):
    for j in range(5):
        ax[i, j].xaxis.set_major_locator(plt.NullLocator())
        ax[i, j].yaxis.set_major_locator(plt.NullLocator())
        ax[i, j].imshow(faces[10*i+j], cmap='bone')
```

```python
x = np.linspace(0, 10, 50)
y1 = np.exp(x)
y2 = np.sin(x)

fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'b')
ax1.set_xlabel('axis x', fontsize=20)
ax1.set_ylabel('exp', color='b', fontsize=22)
ax1.tick_params(axis='y', which='major', color='b')

ax2=ax1.twinx()
ax2.plot(x, y2, 'r')
ax2.set_ylabel('sin', color='r', fontsize=22)
ax2.tick_params(axis='y', color='r', direction='inout')
```

```python
import sys
import numpy as np
import matplotlib.pyplot as plot
from matplotlib.ticker import AutoMinorLocator

x = np.linspace(0, 10, 50)
y1 = np.exp(x)
y2 = np.sin(x)

fig, ax1 = plt.subplots()
ax1.plot(x, y1, 'b')
ax1.set_xlabel('axis x', fontsize=20)
ax1.set_xlim(0, 10)
ax1.set_xticks(range(11))
ax1.set_ylabel('exp', color='b', fontsize=22)

ax1.set_ylim(0, 3000)
ax1.set_yticks(range(0, 3001, 600))

ax1.yaxis.set_minor_locator(AutoMinorLocator(5))
ax1.tick_params(axis='y', which='major', colors='b', direction='in')


ax2=ax1.twinx()
ax2.plot(x, y2, color='r')
ax2.set_ylim(-1.2,1.2)
ax2.set_ylabel('sin', color='r', fontsize=22)
ax2.tick_params(axis='y', color='r', direction='inout')
```

```python
from matplotlib.gridspec import GridSpec
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']

fig=plt.figure()
fig.set_facecolor('w')

fig.suptitle('')

gs = GridSpec(3, 3, figure=fig)

ax1=fig.add_subplot(gs[0, :])
ax2=fig.add_subplot(gs[1, :-1])
ax3=fig.add_subplot(gs[1:, -1])
ax4=fig.add_subplot(gs[-1, 0])
ax5=fig.add_subplot(gs[-1, -2])

def format_axes(fig):
    for i, ax in enumerate(fig.axes):
        ax.text(0.5,0.5, 'ax%d' %(i+1), va='center', ha='center')
        ax.tick_params(labelbottom=False, labelleft=False)

format_axes(fig)
```
