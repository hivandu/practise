```python
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import KMeans

# 输入数据
data = pd.read_excel('data/asia_football_club.xlsx',encoding='gbk')
print(data)
tx = data[['2019年国际排名','2018世界杯','2015亚洲杯']]
df = pd.DataFrame(tx)

# 将数据进行规范化
mms = preprocessing.MinMaxScaler()
tx = mms.fit_transform(tx)

#kmeans算法
kmeans = KMeans(n_clusters=3)
kmeans.fit(tx)
predict_y = kmeans.predict(tx)
print(predict_y)

# 合并聚类结果，插入到元数据中
result = pd.concat((data, pd.DataFrame(predict_y)), axis=1)
result.rename({0:u'聚类'},axis=1,inplace=True)
result
```

```python
# 划分成5类
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import KMeans

# 输入数据
data = pd.read_excel('data/asia_football_club.xlsx',encoding='gbk')
print(data)
tx = data[['2019年国际排名','2018世界杯','2015亚洲杯']]
df = pd.DataFrame(tx)

# 将数据进行规范化
mms = preprocessing.MinMaxScaler()
tsx = mms.fit_transform(tx)

#kmeans算法
kmeans = KMeans(n_clusters=5)
kmeans.fit(tsx)
predict_y = kmeans.predict(tsx)
print(predict_y)

# 合并聚类结果，插入到元数据中
result = pd.concat((data, pd.DataFrame(predict_y)), axis=1)
result.rename({0:u'聚类'},axis=1,inplace=True)
result
```
