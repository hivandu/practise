```python
# 1. 添加数据
import pandas as pd
import numpy as np
from pandas import Series,DataFrame
# # 增加数据
# df = pd.DataFrame([['Snow','M',22],['Tyrion','M',32],['Sansa','F',18],['Arya','F',14]],columns=['name','gender','age'])
# # 添加一列数据
# df['score'] = [90,80,70,60]
# print(df)
# # 添加一行
# new = pd.DataFrame({'name':'lisa','gender':'F','age':19,'score':50},index=[0])
# print(new)
# df = df.append(new,ignore_index=True) # ignore_index=True: 忽略新索引，按原有索引递增
# print(df)

# # 创建一个Series, 用来增加进数据框的最后一行
# new1 = Series([1,2,3,4], index=['name','gender','age','sex'])
# df.append(new1,ignore_index=True)
# print(new1)

# 总结:常用于多个df行方向上进行拼接的，列方向上取并集，没有值则显示NaN

# 删除数据
df = DataFrame(np.arange(9).reshape(3,3),columns=['one','two','three'])
print(df)

# 删除行
df.drop([0,1]) # 删除多行，传入索引列表

# 删除列
df.drop(['one'],axis=1) # 删除列时 axis轴必须指定，否则会报错
```

```python
# 2. 数据的合并-concat
df1 = pd.DataFrame(np.arange(6).reshape(3,2),columns=['four','five'])
df2 = pd.DataFrame(np.arange(6).reshape(2,3),columns=['one','two','three'])
print(df1)
print(df2)

result = pd.concat([df1,df2],axis=0,ignore_index=True,join='outer') # 默认和 df1.append(df2) 相同，行拼接，列进行并集

print(result)

# join='inner' 行进行拼接，取交集

# pd.concat([df1,df2],axis=0,ignore_index=True,join='inner')

# 基于列进行合并
result = pd.concat([df1,df2],axis=1)

print(result)
```

```python
# 3. 数据的合并-join
df3 = pd.DataFrame({'red':[1,3,5],'green':[5,0,3]},index=list('abc'))
df4 = pd.DataFrame({'blue':[1,9,8],'yellow':[6,6,7]},index=list('cde'))
print(df3)
print(df4)

# 参数how：left,right,outer,inner

result = df3.join(df4,how='left')
result = df3.join(df4,how='right')
result = df3.join(df4,how='outer')
result = df3.join(df4,how='inner')
result

# 总结：join常用于数据的列方向的拼接，how= 保留类型
```

```python
# 4. 数据的合并-merge
df1 = pd.DataFrame({'Name':list('ABCDE'),'Sex':['M','F','M','M','F'],'Title':['副教授','讲师','助教','教授','助教']},index=range(1001,1006))
print(df1)
df2 = pd.DataFrame({'Name':list('ABDAX'),'Course':['C++','计算机导论','汇编','数据结构','马克思原理'],'Title':['副教授','讲师','教授','副教授','讲师']},index=[1001,1002,1004,1001,3001])
print(df2)

pd.merge(df1,df2)

# 总结：合并数据的时候需要明确两个数据之间使用的关联字段是什么，例如：『 # 5. 练习合并』
```

```python
# 5. 练习合并
df_info = pd.read_excel('./file/学生信息表.xlsx')
df_score = pd.read_excel('./file/学生成绩表.xlsx')

# 指定合并依据 on

# 假定两边数据口径不一致，一个是“学号”，一个是“学号”
pd.merge(df_info,df_score,left_on='学号', right_on='学号')
```

```python
# 拓展案例练习『将一个csv拆分成几个小的csv』
data = pd.read_csv('./file/ratings_all.csv')

# 分给5个人
person = ['Luobo','JiaNong','JinChun','Fan','Hivan']

# 平分
size = data.shape[0] / len(person)
if data.shape[0] % len(person) != 0:
    size = np.ceil(size).astype(int) # 向上取整，并转成整型
size

# data.iloc[begin:end]
# 1. begin: 0, end: size
# 2. begin: size, end: size+size

for index,use_name in enumerate(person): # 分给每一个人的数据
    begin = index*size
    end = begin+size
    df_sub = data.iloc[begin:end]
    
    # 需要保存csv文件
    file_path = (f'./file/expend/ratings_{use_name}.csv')
    df_sub.to_csv(file_path,index=False)
    
```

```python
# 6. 空值处理
df = pd.read_excel('./file/student_excel.xlsx',header=2)

# 判断每一个值是否为空
# df.isnull()

# 可以得到每一列中有空值的情况
# df.isnull().any()

# 判断一列中全部为空值的情况
# df.isnull().all()

# how的值'all' 表示全部为空值, 'any'只要有空值
df.dropna(axis=1,how='all',inplace=True)

# 通过转置，判断每一行的空值的情况
# df.isnull().T.any()
# df.isnull().T.all()
df.dropna(axis=0,how='all',inplace=True)

# 填充
# df.fillna(0) #把所有空值填写为0
# df.fillna({'分数':0,'姓名':'NaN'},inplace=True)
# df.fillna(method='ffill',inplace=True)

# ffill 前面/左边值填充，bfill后面/右边值填充
# axis 选择制定方向
# df.fillna({'姓名':method='ffill', '分数':0}, inplace=True)

df.fillna({'分数':0},inplace=True)
df.fillna(method='ffill',inplace=True)
df
```

```python
# 7. 去重
df1 = pd.DataFrame({'A':[1,1,1,2,2,3,1],'B':list('aabbbca')})
print(df1)

# 判断每一行是否重复（结果是bool值,TRUE代表重复的）
print("*"*30)
print(df1.duplicated())

# 去除全部的重复行
print("*"*30)
print(df1.drop_duplicates())

# 指定列去处重复行
print("*"*30)
print(df1.drop_duplicates(['A']))

# 保留重复行中的最后一行
print("*"*30)
# Keep参数: last, first, false
print(df1.drop_duplicates(['A'],keep='last'))
```
