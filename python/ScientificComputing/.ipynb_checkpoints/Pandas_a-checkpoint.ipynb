```python
# 8. DataFrame数据获取
df = pd.read_csv('./file/ratings.csv')
print(df)

# 获取一列
df['rating'] # 拿出的结果是series
df[['rating']] # 拿出的是DataFrame
# 获取多列
# df[['rating','movieId']]

# 获取一行 df[] 可以使用切片，可以使用布尔列表
df.index = list('abcde')
print(df['a':'a'])

# df[[True,False,True,False,True]]

# 将评分大于等于5的数据
# df[[i>=5 for i in df['rating']]]
# 简化
df[df['rating'] >= 5]

'''
df.loc[] 通过标签索引获取数据
df.iloc[] 通过位置索引获取数据
'''

# 一个值
df.loc['a','rating']

# 一行
df.loc['a',:]
df.loc['a',['movieId','rating']]
df.loc[['a','b'],['movieId','rating']]

# 选取rating==4的数据
df.loc[df['rating']==4, ['movieId']] # 获取=4的movieId这一列

df.iloc[[0,1,3],:]

```

```python
# 练习，获取9月份数据（不能使用时间模块）
df = pd.read_csv('./file/BJ_tianqi.csv')
df

# 处理数据
## 获取字符串第6位为9的日期抽取出来名（列表生成式）

result = df[[i[5] == '9' for i in df['日期']]] # 第一种
## 
result = df[[i[0:6] == '2019/9' for i in df['日期']]] # 第二种
result

# 老师做法

df2 = pd.read_csv('./file/BJ_tianqi.csv')

# 第一种做法
df2.set_index('日期',inplace=True)
result = df2.loc['2019/9/1':'2019/9/30'] # 但是这个需要时间按顺序排列

# 第二种做法

df2.loc[[i.startswith('2019/9') for i in df2.index.tolist()],:]

# 如果可以用时间模块
df3 = pd.read_csv('./file/BJ_tianqi.csv')
df3['日期'] = pd.to_datetime(df3['日期'])
# df3.info() # 确定数据信息，包括类型
df3.set_index('日期',inplace=True)

# df3.loc['2019/09/01']

# df3.loc['2019/09']

#时间段截取

df3.loc['2019/09/01':'2019/09/07']

# 时间获取（拓展）
df4 = pd.read_csv('./file/BJ_tianqi.csv', parse_dates=['日期'])

df4['月份'] = df4['日期'].dt.month
df4['季度'] = df4['日期'].dt.to_period('Q')
df4
```
