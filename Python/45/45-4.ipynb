```python
# scikit learn

# Min-max规范化
from sklearn import preprocessing
import numpy as np

# 初始化数据,每一行表示一个样本，每一列表示一个特征
x = np.array([[0.,-3.,1.],[3.,1.,2.],[0.,1.,-1.]])
# 进行[0,1]规范化
min_max_scaler = preprocessing.MinMaxScaler()
minmax_x = min_max_scaler.fit_transform(x)
print('**'*20,'\n',minmax_x)

# Z-Score规范化
scaled_x = preprocessing.scale(x)
print('**'*20,'\n',scaled_x)

# 小数定标规范化
j = np.ceil(np.log10(np.max(abs(x))))
scaled_x = x/(10**j)
print('**'*20,'\n',scaled_x)
```

```python
# 可视化图
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 数据准备
N = 1000
x = np.random.randn(N)
y = np.random.randn(N)
# 用Matplotlib画散点图
plt.scatter(x,y,marker='x')
plt.show()
# 用seaborn画散点图
df = pd.DataFrame({'x':x, 'y':y})
sns.jointplot(x='x',y='y',data=df,kind='scatter');
plt.show()
```

```python
# 折线图
# 数据准备
x = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
y = [5, 3, 6, 20, 17, 16, 19, 30, 32, 35]
# Matplotlib
plt.plot(x,y)
plt.show()
# seaborn
df = pd.DataFrame({'x':x,'y':y})
sns.lineplot(x='x',y='y',data=df)
plt.show()
```

```python
# 直方图
a= np.random.randn(100)
s = pd.Series(a)
# Matplotlib
plt.hist(s)
plt.show()

# Seaborn
sns.distplot(s, kde=False)
plt.show()
sns.distplot(s, kde=True)
plt.show()
```

```python
# 条形图
# 数据准备
x = ['Cat1', 'Cat2', 'Cat3', 'Cat4', 'Cat5']
y = [5, 4, 8, 12, 7]

# 用Matplotlib
plt.bar(x,y)
plt.show()

# Seaborn
sns.barplot(x,y)
plt.show()
```

```python
# 箱线图
# 数据准备
# 生成10*4维度数据
data = np.random.normal(size=(10,4))
lables = ['A','B','C','D']

# Matplotlib
plt.boxplot(data, labels=lables)
plt.show()

# Seaborn
df = pd.DataFrame(data, columns=lables)
sns.boxplot(data=df)
plt.show()
```

```python
# 热力图
# 准备数据
## 这里我们使用 Seaborn 中自带的数据集 flights，该数据集记录了 1949 年到 1960 年期间，每个月的航班乘客的数量。
flights = sns.load_dataset('flights')
data = flights.pivot('year','month','passengers')
# Seaborn
sns.heatmap(data)
plt.show()
```

```python
# 蜘蛛图
from matplotlib.font_manager import FontProperties
from matplotlib import font_manager
# 支持中文
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']
# 数据准备
labels = np.array([u'推进','KDA','生存','团站','发育','输出'])
stats = [83,61,95,67,76,88]
# 画图数据准备，角度、状态值
angles = np.linspace(0,2*np.pi, len(labels), endpoint=False)
stats = np.concatenate((stats,[stats[0]]))
angles = np.concatenate((angles,[angles[0]]))

# Matplotlib
fig = plt.figure()
ax = fig.add_subplot(111,polar=True)
ax.plot(angles, stats, 'o-', linewidth=1)
ax.fill(angles, stats, alpha=0.25)

# 设置中文字体
# font = FontProperties(fname='/System/Library/Fonts/PingFang.ttc',size=14)
ax.set_thetagrids(angles * 180/np.pi, labels)
plt.show()
```

```python
# 二元变量分布
import matplotlib.pyplot as plt
import seaborn as sns
# 数据准备
# Seaborn 中自带的数据集 tips， 这个数据记录了不同顾客在餐厅的消费账单及消费情况
# total_bill保存账单金额，tip是给出的消费金额
tips = sns.load_dataset('tips')
print(tips.head(10))
# 用Seaborn画二元变量分布图(散点图，核密度图，Hexbin图)
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind="scatter")
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind="kde")
sns.jointplot(x='total_bill', y = 'tip', data=tips, kind="hex")
plt.show()
```
