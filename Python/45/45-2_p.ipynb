```python
# get net info
def getHTMLText(url):
    try:
        r=requests.get(url)
        r.raise_for_status()
        r.encoding=r.apparent_encoding
        return r.text
    except Exception as err:
        print(err)
```

```python
def getData(html):
    soup=BeautifulSoup(html, 'html.parser')
    ChatList=soup.find('div', attrs={'id':'list_view'})
    dataList=[]
    for Chat in ChatList.find_all('div', attrs={'class':'item_info_view'}):
        data = []
        chatTitle=Chat.find('h2', attrs={'class':'info_title user_select_text'}).getText()
        print(chatTitle)
        data.append(chatTitle)

        # user sum
        bookingNum = Chat.find('span', attrs={'class':'info_num'})
        bookingNum=str(bookingNum.getText().lstrip().split(' ')[0])
        data.append(int(bookingNum))

        # author name
        authorName=Chat.find('span', attrs={'class':'author_title'}).getText()
        if('·' in authorName):
            authorName=authorName.split('·')
        else:
            authorName=authorName.split(' ')
        authorName=authorName[0]
        data.append(authorName)

        # author info
        authorDesc=Chat.find('span', attrs={'class':'author_title'}).getText()
        if('·' in authorDesc):
            authorDesc=authorDesc.split('·')
        else:
            authorDesc=authorDesc.split(' ')
        print(authorDesc)
        authorDesc=authorDesc[1]
        data.append(authorDesc)
        dataList.append(data)
    return dataList
```

```python
def saveData(dataList, path):
    styleBlueBkg = xlwt.easyxf('pattern: pattern solid, fore_colour pale_blue; font: bold on;')

    book=xlwt.Workbook(encoding='utf-8', style_compression=0)

    sheet=book.add_sheet('ChatTop 20', cell_overwrite_ok=True)

    titleList=('Chat标题', '已订阅', '作者', '作者简介')

    first_col=sheet.col(0)
    first_col.width=256*40

    for i in range(0,4):
        sheet.write(0, i, titleList[i], styleBlueBkg)
    
    for i in range(0, len(dataList)):
        data=dataList[i]
        for j in range(0, 4):
            sheet.write(i+1, j, data[j])
    book.save(path)
```

```python
import os
path = os.path.expanduser('~/data/')
chatUrl = 'http://gitbook.cn/gitchat/columns'
html=getHTMLText(chatUrl)
dataList=getData(html)
saveData(dataList, str(path + 'topChat.xls'))
```
