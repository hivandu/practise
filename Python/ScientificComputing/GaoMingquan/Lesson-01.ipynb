## K-Nearest-Neighbors:KNN K近邻

```python
def knn(query_age, data, top_n=3):
    most_similar = sorted(data.items(), key=lambda x_y: (x_y[0] - query_age)**2)[:top_n]
    print(most_similar)
    similary_ys = [y for x, y in most_similar]
    
    return np.mean(similary_ys)
```

## Loss


$$ loss(k, b) = \frac{1}{n} \sum_{i \in X}((y_i - (k * x_i + b))^2$$

```python
import random
```

```python
X_rm = dataframe['RM']
Y = dataframe['price']
```

```python
def loss(k, b):
    return np.mean((Y - (k * X_rm + b))**2)
```

```python
min_loss = float('inf')
best_k, best_b = None, None

for step in range(1000):
    k, b = random.randint(-100, 100), random.randint(-100, 100)
    current_loss = loss(k, b) 
    if current_loss < min_loss:
        min_loss = current_loss
        print('在第{}步，我得到了一个更好的Loss:{}，此时此刻，我的K和B是{}, {}'.format(step, min_loss, k, b))
        best_k, best_b = k, b
```

```python
def partial_k(k, b, X, y):
    return 2 * np.mean((y - (k * X + b)) * (-X))

def partial_b(k, b, X, y):
     return 2 * np.mean((y - (k * X + b)) * (-1))
```

```python
k = 10; b = -10
```

```python
# k_history = []
# b_history = []

min_loss = float('inf')
best_k, best_b = None, None
k, b = random.randint(-50, 50), random.randint(-50, 50)
learning_rate = 1e-03
k_history = []
b_history = []

for step in range(1000):
    k, b = k + -1 * partial_k(k, b, X_rm, Y) * learning_rate, b + -1 * partial_b(k, b, X_rm, Y) * learning_rate
    current_loss = loss(k, b) 
    if current_loss < min_loss:
        min_loss = current_loss
        print('在第{}步，我得到了一个更好的Loss:{}，此时此刻，我的K和B是{}, {}'.format(step, min_loss, k, b))
        best_k, best_b = k, b
        k_history.append(k)
        b_history.append(b)
```

```python
plt.scatter(k_history, b_history)
```
