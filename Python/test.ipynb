```python
import cv2
import matplotlib.pyplot as plt
import os

path = os.path.expanduser('~/data/')

cv2.__version__


# 图像加载

img = cv2.imread(path + "/CV/lena.jfif")

print(img.shape)
plt.imshow(img) #BGR
plt.show()
# 由BGR转换为 RGB
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
canny1 = cv2.Canny(img2, 80, 150)
plt.imshow(canny1)
```

```python
canny2 = cv2.Canny(img2, 50, 100)
plt.imshow(canny2)
```

```python
import cv2
import matplotlib.pyplot as plt

 # 读取图片
img = cv2.imread(path + "/CV/hero.jfif")
# 由BGR转换为 RGB
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
```

```python
plt.figure(figsize=(8,8))
canny1 = cv2.Canny(img, 80, 150)
plt.imshow(canny1)
```

```python
# Soble算子
plt.figure(figsize=(5, 5))
contour = cv2.Sobel(img, -1, 0, 1)
plt.imshow(contour)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 自适应阈值分割
thresh = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV,15,3)
# 查看二值化结果
#cv2.imshow("thres", thresh)
plt.imshow(thresh)
```
