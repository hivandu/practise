```python
# 读取本地文件
def read_file(filename):
    if os.path.exists(filename) is False:
        raise FileNotFoundError('%s not exists'%(filename,))
    f = open(filename,encoding='utf-8')
    content = f.read()
    f.close()
    return content

read_file('./file/file.txt')
```
