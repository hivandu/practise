# Relu 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 绘制relu图
def plot_relu():
    # 设置参数x（起点，终点，间距）
    x = np.arange(-8, 8, 0.2)
    y = relu(x)
    plt.plot(x, y)
    plt.show()
```

```python
# relu函数
def relu(x):
    y = np.where(x<0,0,x)
    return y
```

```python
print(relu(-8))
plot_relu()
```

# sigmod

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 绘制sigmoid图
def plot_sigmoid():
    # 设置参数x（起点，终点，间距）
    x = np.arange(-8, 8, 0.2)
    y = sigmoid(x)
    plt.plot(x, y)
    plt.show()
```

```python
# 绘制sigmoid函数的导数
def plot_ds():
    # 设置参数x（起点，终点，间距）
    x = np.arange(-8, 8, 0.2)
    y = derivative_sigmoid(x)
    plt.plot(x, y)
    plt.show()
```

```python
# sigmoid函数
def sigmoid(x):
    s = 1 / (1 + np.exp(-x))
    return s
```

```python
# sigmoid函数导数
def derivative_sigmoid(x):
    s = 1 / (1 + np.exp(-x))
    ds = s*(1-s)
    return ds
```

```python
print(sigmoid(4))
plot_sigmoid()
plot_ds()
```

# Tanh

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 绘制tanh图
def plot_tanh():
    # 设置参数x（起点，终点，间距）
    x = np.arange(-8, 8, 0.2)
    y = tanh(x)
    plt.plot(x, y)
    plt.show()
```

```python
# tanh函数
def tanh(x):
    y=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    return y
```

```python
print(tanh(4))
plot_tanh()
```
